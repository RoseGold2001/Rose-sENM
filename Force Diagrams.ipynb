```python
import matplotlib.pyplot as plt

# Function to calculate midpoint of vectors
def midpoint(v):
    return [sum(i) / 2 for i in zip(*v)]

# Function to plot force vectors
def plot_forces(tension, gravity, coulomb):
    fig, ax = plt.subplots()
    ax.set_aspect('equal')

    # Plotting the tension force with annotation
    ax.quiver(0, 0, tension[0], tension[1], angles='xy', scale_units='xy', scale=1, color='Purple' , label = 'Tension' , alpha = 0.4)
    tension_mid = midpoint([(0, 0), tension])
    ax.annotate('Tension', xy=(tension_mid[0], tension_mid[1]), xytext=(tension_mid[0] + 0.5, tension_mid[1] + 0.5),
                arrowprops=dict(facecolor='blue', arrowstyle='->'), fontsize=8)

    # Plotting the gravity force with annotation
    ax.quiver(0, 0, gravity[0], gravity[1], angles='xy', scale_units='xy', scale=1, color='green' , label = 'Gravity' , alpha = 0.4)
    gravity_mid = midpoint([(0, 0), gravity])
    ax.annotate('Gravity', xy=(gravity_mid[0], gravity_mid[1]), xytext=(gravity_mid[0] + 0.5, gravity_mid[1] + 0.5),
                arrowprops=dict(facecolor='green', arrowstyle='->'), fontsize=8)

    # Plotting the Coulomb force with annotation
    ax.quiver(gravity[0], gravity[1], coulomb[0], coulomb[1], angles='xy', scale_units='xy', scale=1, color='Red', label = 'Coulomb' , alpha = 0.4)
    coulomb_mid = midpoint([gravity, tension])
    ax.annotate('Coulomb', xy=(coulomb_mid[0], coulomb_mid[1]), xytext=(coulomb_mid[0] + 0.5, coulomb_mid[1] + 0.5),
                arrowprops=dict(facecolor='red', arrowstyle='->'), fontsize=8)

    ax.set_xlim(-10, 10)  # Adjust the limits as per your requirement
    ax.set_ylim(-10, 10)  # Adjust the limits as per your requirement
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.legend()
    plt.title('Force Diagram')
    plt.grid()
    plt.show()

# Example forces acting on an object
tension = (5, -5)  # Tension force in the format (x_component, y_component)
gravity = (0, -5)  # Gravity force in the format (x_component, y_component)
coulomb = (5, 0)  # Coulomb force in the format (x_component, y_component)

plot_forces(tension, gravity, coulomb)
```

```python
import matplotlib.pyplot as plt

# Define charges and their positions
q1 = 3e-6  # Charge in Coulombs
q2 = 12e-6  # Charge in Coulombs
q_test = -2e-9  # Test charge in Coulombs
distance = 1  # Distance between fixed charges in meters

# Define positions of fixed charges
x1, y1 = 0, 0
x2, y2 = distance, 0

# Define positions for the test charge locations
locations = [
    ("Halfway between", (distance / 2, 0)),
    ("Half a meter to the left of +3uC", (x1 - 0.5, y1)),
    ("Half a meter above +12uC", (x2, y2 + 0.5))
]

# Create the figure
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the fixed charges
plt.scatter(x1, y1, color='red', s=100, label='+3uC')
plt.scatter(x2, y2, color='blue', s=200, label='+12uC')

# Plot the test charge at different locations and calculate the electric field
for loc, (x, y) in locations:
    plt.scatter(x, y, color='green', s=50, label='Test Charge')
    dx1, dy1 = x - x1, y - y1
    dx2, dy2 = x - x2, y - y2

    r1 = ((dx1) ** 2 + (dy1) ** 2) ** 0.5
    r2 = ((dx2) ** 2 + (dy2) ** 2) ** 0.5

    k = 8.99e9  # Coulomb's constant
    force1 = k * abs(q1 * q_test) / r1 ** 2
    force2 = k * abs(q2 * q_test) / r2 ** 2

    # Plot the electric field direction (normalized vectors)
    ax.arrow(x, y, -dx1 / r1, -dy1 / r1, head_width=0.1, head_length=0.1, fc='black', ec='black')
    ax.arrow(x, y, -dx2 / r2, -dy2 / r2, head_width=0.1, head_length=0.1, fc='black', ec='black')

    # Show calculated forces for each location
    print(f"{loc}: Force on test charge = {force1 + force2:.4f} N, Direction = Away from the charges")

# Set plot parameters
plt.xlabel('X-axis (m)')
plt.ylabel('Y-axis (m)')
plt.title('Electric Field Lines and Test Charge Positions')
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plotting the charges and position of the test charge
fig, ax = plt.subplots()
ax.set_aspect('equal')

# Plotting the +3μC charge at (0, 0)
ax.scatter(0, 0, color='red', label='+3μC')

# Plotting the +12μC charge at (1, 0)
ax.scatter(1, 0, color='blue', label='+12μC')

# Plotting the test charge at (0.5, 0)
ax.scatter(0.5, 0, color='green', label='-2nC')

# Calculating and plotting the net force vector (in this case, there is no Coulomb force as the test charge is equidistant)
net_force_x = 12 - 3  # As charges are equal distances away, net force is just the difference in magnitudes
ax.arrow(0.5, 0, net_force_x, 0, head_width=0.1, head_length=0.1, fc='black', ec='black', label='Net Force')

# Setting labels and title
ax.set_xlabel('X-axis (m)')
ax.set_xlim (-0.5, 2)
ax.set_ylabel('Y-axis (m)')
ax.set_ylim(-0.2,0.2)
plt.title('Net Force at Halfway Between Charges')

# Showing legend and grid
plt.legend()
plt.grid()
plt.show()
```
